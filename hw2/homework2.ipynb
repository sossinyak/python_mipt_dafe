## Теоретическая справка.

Метод наименьших квадратов - простейший регрессионный метод, позволяющий восстановить исходную функцию по заданному множеству точек. Обычно вид восстанавливаемой зависимости является априорной информацией. В данном задании предполагается, что исходная зависимость - линейная функция вида: $f(x) = ax + b$.

### Постановка задачи.

Заданы два множества: множество абсцисс эксперементальных точек $X = \{x_1, x_2, ..., x_N \} \subset \mathbb{R}$, множество ординат точек $Y = \{y_1, y_2, ..., y_N\}\subset \mathbb{R}$;  
Известно, что существует некоторая зависимость: $f: X \rightarrow Y$, имеющая вид: $f(x) = ax + b$;  
Наша задача состоит в наилучшем определении значений коэффициентов $a$, $b$. 

Под наилучшим подразумеваются такие значения $a$, $b$, что $\sum_{i = 1}^{N}{(y_i - f(x_i))^2} \rightarrow min_{a, b}$

### Алгоритм решения.

Метод нименьших квадратов сводится к вычислению неизвестных параметов по следующим формулам:

$$a = \frac{<xy> - <x><y>}{<x^2> - <x>^2}$$

$$b = <y> - a<x>$$

где $<...>$ - среднее значение.

Данные формулы получены путем решения поставленной задачи оптимизации (вычисления минимума функции  $\sum_{i = 1}^{N}{(y_i - f(x_i))^2}$).
